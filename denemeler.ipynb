```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# Load the obesity dataset
X = df.drop('NObeyesdad', axis=1)
y = df['NObeyesdad']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define a list of classification models
models = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    KNeighborsClassifier(),
    SVC(),
    MLPClassifier(),
]

# Train and evaluate each model
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(type(model).__name__)
    print(classification_report(y_test, y_pred))
    print()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate the evaluation metrics for each model
metrics = []
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    report = classification_report(y_test, y_pred, output_dict=True)
    metrics.append({
        'model': type(model).__name__,
        'accuracy': report['accuracy'],
        'recall': report['macro avg']['recall'],
        'f1-score': report['macro avg']['f1-score'],
        'precision': report['macro avg']['precision'],
    })

# Convert the metrics list into a Pandas DataFrame
df = pd.DataFrame(metrics)

# Create a bar chart showing the evaluation metrics for each model
fig, ax = plt.subplots()
x = np.arange(len(models))
accuracy = df['accuracy']
recall = df['recall']
f1 = df['f1-score']
precision = df['precision']
ax.bar(x - 0.2, accuracy, 0.2, label='accuracy')
ax.bar(x, recall, 0.2, label='recall')
ax.bar(x + 0.2, f1, 0.2, label='f1-score')
ax.bar(x + 0.4, precision, 0.2, label='precision')
ax.set_xticks(x)
ax.set_xticklabels(df['model'])
ax.legend()
plt.xticks(rotation=90)
plt.show()
```
